```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.loadtxt('/content/hbonds.dat')  # Add skiprows=1 if there's a header

# Extract columns
x = data[:, 0]
y = data[:, 1]

# Convert X-axis to time in ns
x_time = x / 10

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x_time, y, label='H-bond Frequency', color='blue')

# Labels and title
plt.xlabel('Time (ns)')
plt.ylabel('Frequency of H-bond')
plt.title('H-bond Frequency Over Time')

# Y-axis ticks: only 0, 1, 2
plt.yticks([0, 1, 2])

# Remove background grid
plt.grid(False)

# Optional: Remove top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Save figure at 600 DPI
plt.savefig('3aup_hbond_frequency.png', dpi=600, bbox_inches='tight')

# Show plot
plt.show()
```
